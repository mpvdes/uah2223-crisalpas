## Carga de datos

Crear dos dataframes:

* `hoteles`: recoge la información de los hoteles. Fichero `hoteles.csv`
* `reviews`: recoge la información de las revisiones de todos los hoteles en todos los años de los que disponemos. Ficheros `reviews_20XX.xlsx`. Repasa la operación `p.concat` de la librería `pandas`. Esto te permitirá construir un único dataframe a partir de una lista de dataframes. La carga de reviews puede tardar un poco.

````python
# Sol
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os  # librería del sistema operativo, útil para mergear distintos archivos con nombre similar
import glob as glob
````

````python
hoteles = pd.read_csv('datos/hoteles.csv')
````

Aunque en el caso del csv de hoteles únicamente tenemos uno, en el caso de las reviews hay varios archivos. En el caso de tener que hacer merge en este tipo de datos, usamos la librería `os`.

````python
ficheros = os.listdir('datos/reviews') # guardamos una lista con todos los archivos dentro de esa carpeta, que son los que queremos abrir
ficheros
````

````python
tablas = [] # creamos una lista vacía
for f in ficheros:
    print('datos/reviews/' + f) # vemos los archivos que nos está ejecutando
    t = pd.read_excel('datos/reviews/' + f) # + f lo que hace es 'sumarnos' un archivo y, por lo tanto, darnos el siguiente
    tablas.append(t) # lee todos los ficheros y lo guarda en la lista que hemos llamado tablas

print(len(tablas))
````

````python
reviews = pd.concat(tablas)   # hacemos el concat de todos los dataframe que hemos leído
````

También se puede hacer usando la librería **glob**, ya que en este caso además tienen el mismo nombre.

````python
xlsx_files = r"datos/reviews"
reviews = pd.DataFrame()
reviews = pd.concat(map(pd.read_excel,glob.glob(os.path.join(xlsx_files,'reviews_20*.xlsx'))), ignore_index= True)
reviews
````

## Reviews. Valoraciones por año

__a)__ Cuántos comentarios tenemos por año. 

Una forma de agrupar datos por periodos de tiempo (dias, meses, años, etc) es usando la operación `resample`. Esta operación permite hacer grupos, al igual que la operación Groupby, teniendo en cuenta que la variable de agrupación es de tipo fecha. Permite agrupar filas con fecha en el mismo mes ('M'), en el mismo año ('A'), etc.
La documentación y ejemplos de uso de dicha función la puedes encontrar en el enlace:  https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html.

Ejemplo de uso:

```
             
>>> tabla
   price  volume          date
0     10      50    2018-01-07
1     11      60    2018-01-14
2      9      40    2018-01-21
3     13     100    2018-01-28
4     14      50    2018-02-04
5     18     100    2018-02-11
6     17      40    2018-02-18
7     19      50    2018-02-25

>>> tabla.resample('M', on='date').mean()

               price  volume
2018-01-31     10.75    62.5
2018-02-28     17.00    60.0
```

__b)__ Dibuja un gráfico donde se vea la evolución del número de comentarios por año. Se trata de representar una serie donde el índice sea la fecha y el valor sea el número de comentarios. El tipo de gráfico (kind) lo puedes dejar vacío.

````python
# Sol a.

comentarios_anual = db_completa.resample('Y', on='date').message.count()
comentarios_anual
````

````python
# Sol b.

figura, g = plt.subplots()  # creo el espacio para la figura

comentarios_anual.plot(ax = g, kind='line') # creo el gráfico
````
