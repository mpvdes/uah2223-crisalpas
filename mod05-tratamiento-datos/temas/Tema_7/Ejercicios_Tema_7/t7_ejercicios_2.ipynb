__EJ 12.__ A√±adir una columna con las menciones a otras cuentas. La idea es extraer partes del texto y mostrarlo como una columna nueva usando expresiones regulares junto con la funci√≥n `findall`.

Ejemplos:

```python
# Extraer una palabra concreta, por ejemplo 'Colombia'
import re

resultado = tweets['texto'].str.findall(r'(Colombia)', flags = re.IGNORECASE)
resultado
```

```python
# Extraer n√∫meros
er = r'([0-9]+)'

resultado = tweets['texto'].str.findall(er, flags = re.IGNORECASE)
resultado
```

```python
# Extraer urls del texto
er = r'(http+[\S]*)'

resultado = tweets['texto'].str.findall(er, flags = re.IGNORECASE)
resultado
```

`\w` significa cualquier car√°cter alfanum√©tico

`\W` significa cualquier car√°cter 

`* ` significa 0 o m√°s repeticiones

`+ ` significa 1 o m√°s repeticiones

```python
# sol:
import re
tweets["menciones"] = tweets['texto'].str.findall(r'([@][\w]*)', flags = re.IGNORECASE)
tweets
```
